#### 1. What is the difference between mean, median, and mode?
#### Real-World Example
    In a company's employee salary data:
    
    Mean: Average salary.
    Median: Middle salary, robust to outliers.
    Mode: Most frequently occurring salary.
    Insight: Outliers (8000 repeated) pull the mean up, while the median reflects a better central tendency for skewed data.

```python
# Example: Salaries of employees
salaries = [3000, 3500, 4000, 4500, 5000, 8000, 8000, 8000]

# Mean
mean = np.mean(salaries)

# Median
median = np.median(salaries)

# Mode
mode = stats.mode(salaries)

print(f"Mean Salary: {mean}")
print(f"Median Salary: {median}")
print(f"Mode Salary: {mode.mode} (Count: {mode.count})")
```

#### 5. What is the difference between population and sample variance?
    Real-World Example
    Population variance: When analyzing the entire population (e.g., all student grades in a school).
    Sample variance: When analyzing a sample (e.g., grades of a classroom).
    Insight:
    Population variance uses ùëÅ and N as denominator.
    Sample variance uses ùëÅ‚àí1 . N‚àí1 to correct bias in small samples.

```python
# Population data
population_data = [2, 4, 4, 4, 5, 5, 7, 9]

# Sample data (subset of population)
sample_data = [2, 4, 4, 5]

# Population Variance
population_variance = np.var(population_data)

# Sample Variance (ddof=1 for unbiased estimator)
sample_variance = np.var(sample_data, ddof=1)

print(f"Population Variance: {population_variance:.2f}")
print(f"Sample Variance: {sample_variance:.2f}")
```

#### 6. When the mean, median, and mode are approximately equal, what type of distribution is it?
    Real-World Example
    Heights of adults typically follow a normal distribution, where mean ‚âà median ‚âà mode.


#### Insight: This is a hallmark of a normal distribution.
    Normal Distribution and Mean, Median, Mode
    In a perfect normal distribution:
    
    Mean = Median = Mode
    This happens because the data is symmetrical and evenly distributed around the center.
    However, in real-world datasets or simulated data, we might observe:
    
    Mean ‚âà Median but Mode differs slightly.
    Why Does Mode Differ?
    Nature of Mode:
    
    The mode represents the most frequently occurring value in a dataset.
    In continuous data (like a normal distribution), values are usually not repeated. As a result, the mode can differ slightly or may not exist as a single unique value.
    Sampling Noise:
    
    When simulating or sampling data from a normal distribution, there might be small deviations due to randomness.
    The mode can appear slightly different because it depends on the most frequent value within the sampled data, not the theoretical center of the distribution.


```python
import matplotlib.pyplot as plt

# Normally distributed data
data = np.random.normal(50, 10, 1000)  # mean=50, std=10

# Calculate mean, median, mode
mean = np.mean(data)
median = np.median(data)
mode = stats.mode(data)

print(f"Mean: {mean:.2f}")
print(f"Median: {median:.2f}")
print(f"Mode: {mode.mode:.2f}")

# Visualize the distribution
plt.hist(data, bins=30, color='lightblue', edgecolor='black')
plt.title("Normal Distribution")
plt.axvline(mean, color='red', linestyle='dashed', linewidth=1, label="Mean")
plt.axvline(median, color='green', linestyle='dashed', linewidth=1, label="Median")
plt.legend()
plt.show()
```

#### 4. How do you interpret the IQR (Interquartile Range), and how is it used to detect outliers?
    Real-World Explanation:
    The IQR measures the spread of the middle 50% of the data:
    
    Outliers are values that lie beyond 
    Q1‚àí1.5√óIQR or Q3+1.5√óIQR.
    Insight: IQR is robust to extreme values, and outliers can be identified easily.

```python

data = [10, 12, 15, 18, 20, 22, 25, 30, 100]  # Outlier present

# Calculate Q1, Q3, and IQR
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1

# Calculate bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Find outliers
outliers = [x for x in data if x < lower_bound or x > upper_bound]

print(f"Q1: {Q1}, Q3: {Q3}, IQR: {IQR}")
print(f"Outliers: {outliers}")
```

#### 8.How would you compare two datasets using measures of central tendency and dispersion?
    Real-World Example: Comparing Salaries of Two Departments
    Suppose you have the monthly salaries of employees in two departments:
    
    Department A has high variability (more spread).
    Department B is more consistent (less spread).
    We compare:
    
    Mean/Median to check central tendency (average salary).
    Standard Deviation and IQR to analyze how spread out the salaries are.
Insight : 
    When comparing two datasets:
    
    Use central tendency (mean, median) to understand where the data is centered.
    Use dispersion (standard deviation, IQR) to measure variability.
    Visualize with boxplots or histograms for intuitive comparison.

```python
# Salaries of employees in two departments
department_A = [3000, 3200, 3500, 4000, 4100, 5000, 7000, 12000]  # High spread
department_B = [3500, 3600, 3700, 3800, 3900, 4000, 4100, 4200]   # Low spread

# Calculate central tendency
mean_A, median_A = np.mean(department_A), np.median(department_A)
mean_B, median_B = np.mean(department_B), np.median(department_B)

# Calculate dispersion
std_A, iqr_A = np.std(department_A), np.percentile(department_A, 75) - np.percentile(department_A, 25)
std_B, iqr_B = np.std(department_B), np.percentile(department_B, 75) - np.percentile(department_B, 25)

# Print comparisons
print("Department A:")
print(f"Mean: {mean_A:.2f}, Median: {median_A}, Std Dev: {std_A:.2f}, IQR: {iqr_A}\n")

print("Department B:")
print(f"Mean: {mean_B:.2f}, Median: {median_B}, Std Dev: {std_B:.2f}, IQR: {iqr_B}")

# Boxplot for Visualization
plt.boxplot([department_A, department_B], labels=["Department A", "Department B"])
plt.title("Salary Comparison: Department A vs. Department B")
plt.ylabel("Salaries")
plt.grid(True)
plt.show()
```

#### 6.What is Kurtosis and How to Interpret It?
    Kurtosis is a statistical measure that describes the shape of the tails in a data distribution, particularly focusing on:
    
    How heavy or light the tails are compared to a normal distribution.
    Whether extreme values (outliers) are likely in the dataset.

#### Types of Kurtosis
    Mesokurtic (Normal Distribution):
    
    Kurtosis ‚âà 3 (Excess Kurtosis ‚âà 0).
    Tails are similar to a normal distribution.
    
    Leptokurtic (Heavy-Tailed):
    
    Kurtosis > 3 (Excess Kurtosis > 0).
    The distribution has heavier tails and a sharper peak.
    Interpretation: Higher likelihood of extreme values (outliers).
    Example: Financial returns where extreme gains or losses occur.
    
    Platykurtic (Light-Tailed):
    
    Kurtosis < 3 (Excess Kurtosis < 0).
    The distribution has lighter tails and a flatter peak.
    Interpretation: Fewer extreme values compared to normal distribution.
    Example: Test scores where most students score near the mean.

#### Real-World Example
    Stock Market Returns:
    Leptokurtic distributions indicate higher chances of extreme price changes.
    Quality Control:
    In manufacturing, platykurtic distributions suggest consistent product measurements with fewer outliers.

#### Key Insight
    High kurtosis ‚Üí Be cautious of outliers and extreme values.
    Low kurtosis ‚Üí The data is stable, with fewer extreme observations.

```python
from scipy.stats import kurtosis

# Generate three distributions
normal_data = np.random.normal(0, 1, 1000)  # Mesokurtic
heavy_tailed_data = np.random.laplace(0, 1, 1000)  # Leptokurtic
light_tailed_data = np.random.uniform(-1, 1, 1000)  # Platykurtic

# Calculate kurtosis
kurt_normal = kurtosis(normal_data)
kurt_heavy = kurtosis(heavy_tailed_data)
kurt_light = kurtosis(light_tailed_data)

# Print kurtosis values
print(f"Normal Distribution Kurtosis: {kurt_normal:.2f} (Mesokurtic)")
print(f"Heavy-Tailed Distribution Kurtosis: {kurt_heavy:.2f} (Leptokurtic)")
print(f"Light-Tailed Distribution Kurtosis: {kurt_light:.2f} (Platykurtic)")

# Plot the distributions
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.hist(normal_data, bins=30, color="lightblue", edgecolor="black")
plt.title("Mesokurtic (Normal Distribution)")

plt.subplot(1, 3, 2)
plt.hist(heavy_tailed_data, bins=30, color="lightgreen", edgecolor="black")
plt.title("Leptokurtic (Heavy-Tailed)")

plt.subplot(1, 3, 3)
plt.hist(light_tailed_data, bins=30, color="lightcoral", edgecolor="black")
plt.title("Platykurtic (Light-Tailed)")

plt.tight_layout()
plt.show()
```

#### How do you test if two datasets have equal variance (homoscedasticity)?
    Testing If Two Datasets Have Equal Variance (Homoscedasticity)
    To test if two datasets have equal variance (homoscedasticity), you can use statistical tests like:
    
    Levene's Test
    Bartlett's Test
    Both tests are used to check the assumption of equal variances, which is crucial for certain statistical tests such as t-tests and ANOVA.
    


#### 1. Levene's Test
    Purpose: Tests the null hypothesis that the variances of two or more datasets are equal.
    Robustness: It is less sensitive to deviations from normality compared to Bartlett's Test.
    Steps:
    If the p-value > 0.05: Fail to reject the null hypothesis ‚Üí Variances are equal.
    If the p-value ‚â§ 0.05: Reject the null hypothesis ‚Üí Variances are not equal.


#### Real-World Example: Compare Variance of Two Datasets (Test Scores)
Suppose you want to check whether the test scores of students from two different classes have equal variances.

```python
import numpy as np
from scipy.stats import levene

# Test scores for two classes
class_A = [85, 88, 90, 92, 95, 100, 105]
class_B = [70, 75, 80, 85, 90, 95, 100]

# Perform Levene's test
stat, p_value = levene(class_A, class_B)

print(f"Levene's Test Statistic: {stat:.2f}")
print(f"P-Value: {p_value:.3f}")

if p_value > 0.05:
    print("Fail to reject the null hypothesis: Variances are equal (Homoscedasticity).")
else:
    print("Reject the null hypothesis: Variances are not equal (Heteroscedasticity).")
```

#### 2. Bartlett's Test
    Purpose: Tests the null hypothesis that the variances of two or more datasets are equal.
    Robustness: Bartlett‚Äôs Test is sensitive to non-normal data.
    Steps:
    If the p-value > 0.05: Fail to reject the null hypothesis ‚Üí Variances are equal.
    If the p-value ‚â§ 0.05: Reject the null hypothesis ‚Üí Variances are not equal.


```python
from scipy.stats import bartlett

# Perform Bartlett's test
stat, p_value = bartlett(class_A, class_B)

print(f"Bartlett's Test Statistic: {stat:.2f}")
print(f"P-Value: {p_value:.3f}")

if p_value > 0.05:
    print("Fail to reject the null hypothesis: Variances are equal (Homoscedasticity).")
else:
    print("Reject the null hypothesis: Variances are not equal (Heteroscedasticity).")
```

#### When to Use Which Test?
    Levene's Test: Preferred when data may not follow a normal distribution.
    Bartlett's Test: Suitable when data is assumed to be normally distributed.


#### Why Does Equal Variance Matter?
    Equal variances (homoscedasticity) are an assumption in statistical tests such as:
    
    t-tests (independent samples).
    ANOVA (comparing means across multiple groups).
    If variances are not equal (heteroscedasticity), you should use alternative methods like:
    
    Welch's t-test instead of a standard t-test.
    Robust statistical methods.
